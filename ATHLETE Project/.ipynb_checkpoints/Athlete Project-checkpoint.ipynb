# Project Athlete finder


```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

## Early EDA/vizualizations

```python
athlete_raw = pd.read_csv('bodyPerformance.csv')
athlete_raw
```

```python
sns.pairplot(athlete_raw[['height_cm',"weight_kg","body fat_%","gender"]],hue = 'gender')
```

```python
male_athlete = athlete_raw[athlete_raw['gender'] == 'M']
female_athlete = athlete_raw[athlete_raw['gender'] == 'F']
```

```python
sns.pairplot(male_athlete[['height_cm',"weight_kg","body fat_%","class","sit-ups counts"]],hue = 'class')
```

```python
sns.pairplot(female_athlete[['height_cm',"weight_kg","body fat_%","class"]],hue = 'class')
```

```python
male_athlete.describe()
```

```python
female_athlete.describe()
```

```python
new_list = []
for i in male_athlete['class']:
    if i == "D" or i == "C": # Bad Athletes
        new_list.append(1)
    else: # Good Athletes 
        new_list.append(0)
male_athlete['bin_class'] = new_list
```

```python
new_list = []
for i in female_athlete['class']:
    if i == "D" or i == "C":
        new_list.append(1)
    else:
        new_list.append(0)
female_athlete['bin_class'] = new_list
# Data is now binary
```

```python
sns.pairplot(male_athlete[['height_cm',"weight_kg","body fat_%","bin_class","sit-ups counts"]],hue = 'bin_class')
```

The good athletes have been put into class 0(blue) and the bad athletes have been put into the class 1(orange). There isn't a really great way to separate the data but this will prove the point of mental capacity coming into play more than pure body measurements alone.

 ## Machine Learning

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import time
```

```python
#Making data into X and y
test_data = male_athlete
X = test_data.drop(['class','bin_class','gender'],axis=1)
y = male_athlete['bin_class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=1)
```

### Multi-layer Perceptron (Neural Network)

```python
mlp = MLPClassifier(solver='sgd', alpha=1e-3,hidden_layer_sizes=(10,20,30),random_state=1).fit(X_train, y_train)
```

```python
mlp.score(X_test, y_test)
```

 We can see here with a score of ~85 the model isn't as accurate as planned. Lets try to use data preprocessing to change that score.

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
pipe = make_pipeline(StandardScaler(), MLPClassifier(activation = "logistic",solver='lbfgs', alpha=1e-3,
                                                     hidden_layer_sizes=(100,50),random_state=1,max_iter=1000))
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

```python
pipe = make_pipeline(StandardScaler(), MLPClassifier(activation = "tanh",solver='sgd', alpha=1e-4,
                                                    hidden_layer_sizes=(100,50),random_state=1,max_iter=1000))
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

```python
pipe = make_pipeline(StandardScaler(), MLPClassifier(activation = "logistic",alpha=1e-3,
                                                     hidden_layer_sizes=(100,50),random_state=1,max_iter=1000))
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

Preprocessing raised the accuracy of the model slightly when sgd as the solver.

## Logistic Regression

A little lower than MLP so now we will now used the scaled data.

```python
#BEST LOG MODEL
pipe = make_pipeline(StandardScaler(), LogisticRegression())
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

Still lower than the MLP model

## Vizualizations

```python
# Getting results for runttimes and score
score = []
time_list = []


# Multi-layer Perceptron
start = time.time()
pipe = make_pipeline(StandardScaler(), MLPClassifier(activation = "tanh",solver='sgd', alpha=1e-4,
                                                    hidden_layer_sizes=(100,50),random_state=1,max_iter=1000))
pipe.fit(X_train, y_train)
end = time.time()
time_list.append(end - start)
score.append(pipe.score(X_test, y_test))
# Logistic Regression
start = time.time()
pipe = make_pipeline(StandardScaler(), LogisticRegression())
pipe.fit(X_train, y_train)
end = time.time()
time_list.append(end - start)
score.append(pipe.score(X_test, y_test))
#Decision Tree Classifier
start = time.time()
pipe = make_pipeline(StandardScaler(), tree.DecisionTreeClassifier())
pipe.fit(X_train, y_train)
end = time.time()
time_list.append(end - start)
score.append(pipe.score(X_test, y_test))
#SVC
start = time.time()
pipe = make_pipeline(StandardScaler(),svm.SVC())
pipe.fit(X_train, y_train)
end = time.time()
time_list.append(end - start)
score.append(pipe.score(X_test, y_test))
#Knearest Neighbor
start = time.time()
pipe = make_pipeline(StandardScaler(),KNeighborsClassifier(n_neighbors=19))
pipe.fit(X_train, y_train)
end = time.time()
time_list.append(end - start)
score.append(pipe.score(X_test, y_test))
#SGD
start = time.time()
pipe = make_pipeline(StandardScaler(),SGDClassifier(tol = 1e-3,random_state=1))
pipe.fit(X_train, y_train)
end = time.time()
time_list.append(end - start)
score.append(pipe.score(X_test, y_test))
```

```python
# Bar Plot making comparisons 
models = ['MLP','Log_R','Decision Tree','SVC','Knn','SGD']
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
sns.barplot(x=models,y=score)
```
